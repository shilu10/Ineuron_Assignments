#### 3. Describe the mathematical operation that a 3x3 kernel performs on a single pixel in an image.

    The Convolution Process involves these steps. (1)It places the Kernel Matrix over each pixel of the image (ensuring that the full Kernel is within the image), multiplies each value of the Kernel with the corresponding pixel it is over. (2)Then, sums the resulting multiplied values and returns the resulting value as the new value of the center pixel. (3) This process is repeated across the entire image.

#### 5. What exactly is padding?

    Padding is to add extra pixels outside the image. And zero padding means every pixel value that you add is zero. 
    Whenever the convolutional operation is happening with kernel and image or feature map, the output of will have a smaller spatial dimension than the input image or feature map, so if we want to preserve the input shape, we will use the padding="same" or padding=1. Which is make sure to have a same input_shape of image and same output_shape of image after the convolutional operation.
    
    When performing a standard convolution operation, the image shrinks by a factor equivalent to the filter size plus one. If we take an image of width and height 6, and a filter of width and height 3, the image shrinks by the following factor.

    eg:
        if image_size is 10 * 10 and filter is 3 * 3, then, 10 - 3 + 1 = 8 
        So, the output of the convolutional operation will be 8 * 8.
        
    We can add a padding of any size, depending on retrieving the image dimension after the convolutional operation.
    
    formula for he padding is,
     padding= f - 1 /2

#### 6. What is the concept of stride?

    The stride simply describes the step size when sliding the convolutional filter over the input image. In the previous examples, we’ve always slid the filter by one pixel rightwards or downwards. We’ve used a stride of 1.
    With a stride of 2, we would slide the window by two pixels on each step.
   <img src="https://programmathically.com/wp-content/uploads/2021/12/Screenshot-2021-12-03-at-12.12.24-1024x913.png" height="15px" />
   
    Generally speaking, the smaller the steps you take when sliding the filter over an image, the more details will be reflected in the resulting feature map. It also means that more features will be shared between the outputs since large portions of the filters will overlap.
    
    If we increase the step size, fewer parameters are shared between filters, and the feature map is smaller. Applying a larger stride basically has the effect of downsampling the image so that lower level details are obscured.

#### 8. What exactly is a channel?

    There are a handful of colour spaces to represent an image such as RGB, BGR, HSV, CMYK etc. But there is something they all have in common. They are the channels, which these colour spaces use, to collectively form an image. 
    
    Every Image is madeup of pixels, and each pixel value is a combination of values of colors, which is known as a primary colors. channel is a grayscale image of the colored image. which will be combined to form a colored image. 
    
    GrayScale image, is different from black and white image, where black and white image, just have a binary values either 0 or 255, Grayscale image, will show the intensity of the light.
    
    when we have a RGB image, each pixel value is represented in combination of 3 colors red, green, blue. Each channel represents the Intesity of red, Intensity of green color, and Intensity of blue color.
    
    When we split up each channel in the RGB image, it will just a 3 grayscale image, shows their color intensity.

```python
import cv2 as cv
import numpy as np
#read the image
image = cv.imread(umbrella_img_path)
#convert the image to RGB (images are read in BGR in OpenCV)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
#split the image into its three channels
(R,G,B) = cv.split(image)

#write the images to disk
cv.imwrite("channel_red.jpg", R)
cv.imwrite("channel_green.jpg", G)
cv.imwrite("channel_blue.jpg", B)
```
